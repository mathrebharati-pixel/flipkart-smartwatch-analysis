```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import requests
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
```

```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import time

Sponsored = [] 
Productname = []
Colour=[]
Fassured = []
Price = []
Original_price = []
Discount = []
Promotion_tag=[]
Ratings=[]
Reviews=[]
pagenum = []


total_time = time.time()

for i in range(1,101): 
    start_time = time.time()
    URL = f"https://www.flipkart.com/search?q=smart+wearable&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&otracker=nmenu_sub_Electronics_0_Smart+Wearable+Tech&page={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.text, "html.parser")

    for x in soup.find_all('div', attrs={'class':'RGLWAk'}):
        span = x.find('div', attrs={'class':'IxWX8O'})
        prod = x.find('a', attrs={'class':'pIpigb'})   
        colo= x.find('div', attrs={'class':'U_GKRr'})
        fass = x.find('div', attrs={'class':'FIdGa1'})
        pr = x.find('div', attrs={'class':'hZ3P6w'})
        op = x.find('div', attrs={'class':'kRYCnD'})
        discount = x.find('div', attrs={'class':'HQe8jr'})
        Promo= x.find('div', attrs={'class':'MaiFhH'})
        ratings= x.find('div', attrs={'class':'MKiFS6'})
        reviews= x.find('span', attrs={'class':'PvbNMB'})
        

        Sponsored.append( True if  span else False)
        Productname.append(prod.text if prod else np.nan)
        Colour.append(colo.text if colo else np.nan)
        Fassured.append(True if fass else False)
        Price.append(pr.text if pr else np.nan)
        Original_price.append(op.text if op else np.nan)
        Discount.append(discount.text if discount else np.nan)
        Promotion_tag.append( Promo.text if  Promo else np.nan)
        Ratings.append(ratings.text if ratings else np.nan)
        Reviews.append( reviews.text if  reviews else np.nan)

        pagenum.append(i)
    print(f'Page {i} completed in {time.time()-start_time:.2f} seconds')

print("Total Time Completed in seconds", str(time.time()-total_time))
```

```python
df=pd.DataFrame({"Productname":Productname,
                 "Colour":Colour,
                 "Flipkart_assured":Fassured,
                 "Sponsored":Sponsored,
                 "Price":Price,
                 "Original_price":Original_price,
                 "Discount":Discount,
                 "Promotion_tag":Promotion_tag,
                 "Page_num" : pagenum,
                 "Ratings":Ratings,
                 "Reviews":Reviews})
```
