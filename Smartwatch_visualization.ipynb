```python
# Univariate
# Brand count
```

```python
sns.countplot(y="Brand", data=df, order=df["Brand"].value_counts().head(10).index, palette='tab10')
plt.title("Top 10 Brands ")
plt.show()
```

```python
df['Strap_color'].value_counts().head(10).plot(kind = 'line')
plt.title('Top 5 strap colors')
plt.show()
```

```python
sns.histplot(x = 'Price', kde = True, data = df)
plt.title('Price Distribution')
plt.show()
```

```python
q1 = df['Price'].quantile(0.25)
q3 = df['Price'].quantile(0.75)
iqr = q3 - q1
outliers = df[(df['Price'] < q1 - 1.5 * iqr) | (df['Price'] > q3 + 1.5 * iqr)]
outliers
```

```python
top_outlier_brands = outliers['Brand'].value_counts().head(5)
plt.figure()
plt.pie(top_outlier_brands, labels=top_outlier_brands.index, autopct='%1.1f%%')
plt.title("Top 5 Brands Contributing to Outliers")
plt.show()
```

```python
# KDE plot is used to visualize the distribution of a continuous variable.
# It provides a smooth estimate of data density, helping to understand where values are concentrated.
# It shows the probability density distribution of ratings.
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['Ratings'], fill=True)
plt.title("Density of Ratings")
plt.show()
```

```python
# Bivariate
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='Price_category', y='Ratings', data=df, jitter=True, palette='Set2')
plt.title('Rating Distribution by Price Category')
plt.xlabel('Price Category')
plt.ylabel('Ratings')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(data=df, x='Price', y='Reviews')

plt.title("Price vs Reviews", fontsize=14, weight='bold')
plt.xlabel("Price")
plt.ylabel("Reviews")
plt.show()
```

```python
top10_brands = df['Brand'].value_counts().head(10).index
df_top10 = df[df['Brand'].isin(top10_brands)]

plt.figure(figsize=(12,6))

sns.violinplot(
    data=df_top10,
    x='Brand',
    y='Ratings',
    order=top10_brands
)

plt.xticks(rotation=45)
plt.title("Top 10 Brands – Ratings Distribution", fontsize=14, weight='bold')
plt.show()
```

```python
sns.countplot(data=df, x='Sponsored', hue='Flipkart_assured')
plt.title("Impact of Sponsorship on Flipkart Assured Product Count")

plt.show()
```

```python
# Multivariate
```

```python
sns.heatmap(df[['Price','Original_price','Ratings','Reviews']].corr(), annot=True)
plt.title("Correlation Between Price, Ratings, and Reviews")
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(
    x="Price_category",
    y="Ratings",
    hue="Sponsored",
    data=df,
    palette="tab10"
)
plt.title("Ratings by Price Category and Sponsored Status", fontsize=14)
plt.show()
```

```python
g = sns.pairplot(
    df,
    vars=['Price_category', 'Ratings', 'Reviews'],
    hue='Price_category',
    diag_kind='hist'
)

g.fig.suptitle("Multivariate Analysis: Price Category vs Ratings vs Reviews",
               y=1.02, fontsize=14)

plt.show()
```
